```python
import pandas as pd
```

```python
df_j = pd.read_csv("./results/test_test_gpt-j_500.csv")
```

```python
def clean_y_hat(y_hat): 
    y_hat = y_hat.lower()
    if ("needs" in y_hat or "need" in y_hat) and "caution" in y_hat: 
        return "__needs_caution__" 
    if ("needs" in y_hat or "need" in y_hat) and "intervention" in y_hat: 
        return "__needs_intervention__" 
    if "casual" in y_hat: 
        return "__casual__" 
    return "IRRELEVANT"
    
df_j.gold = df_j["gold"].replace("needs_caution", "__needs_caution__")
df_j.y_hat = df_j.y_hat.apply(clean_y_hat)
df_j['model'] = 'gpt-j'
df_j
```

```python
gpt_j_acc = len(df_j[df_j.y_hat == df_j.gold])/len(df_j)
```

```python
df_2 = pd.read_csv("./results/test_test_gpt2_500.csv")
df_2.gold = df_2["gold"].replace("needs_caution", "__needs_caution__")
df_2.y_hat = df_2.y_hat.apply(clean_y_hat)
df_2['model'] = 'gpt-2'
gpt_2_acc=len(df_2[df_2.y_hat == df_2.gold])/len(df_2)
df_2
```

```python
df_n = pd.read_csv("./results/test_test_gpt-neo_500.csv")
df_n.gold = df_n["gold"].replace("needs_caution", "__needs_caution__")
df_n.y_hat = df_n.y_hat.apply(clean_y_hat)
df_n['model'] = 'gpt-neo'
gpt_neo_acc=len(df_n[df_n.y_hat == df_n.gold])/len(df_n)
df_n
```

```python
df_3 = pd.read_csv("./results/test_test_gpt-3_500.csv", index_col=0)
df_3.gold = df_3["gold"].replace("needs_caution", "__needs_caution__")
df_3.y_hat = df_3.y_hat.apply(clean_y_hat)
df_3['model'] = 'gpt-3'
gpt_3_acc=len(df_3[df_3.y_hat == df_3.gold])/len(df_3)
df_3
```

```python
accs = {
    'gpt-2':gpt_2_acc,
    'gpt-j':gpt_j_acc,
    'gpt-3':gpt_3_acc,
    'gpt-n':gpt_neo_acc,
}
accs
```

```python
df = pd.concat((df_3, df_2, df_j, df_n))
df = df.drop(columns=['response_obj', 'question'])
df['y_hat'] = df['y_hat'].apply(lambda x: x.replace("__",""))
df['gold'] = df['gold'].apply(lambda x: x.replace("__",""))

import seaborn as sns
import matplotlib.pyplot as plt
df
```

```python
count_preds_by_model_and_category = {}

for m in list(df.model.unique()):
    counts = {}
    for c in df.gold.unique():
        preds = df[df['y_hat'] == c]
        counts[c] = len(preds)
    count_preds_by_model_and_category[m] = counts
# accs_by_model
count_preds_by_model_and_category
```
